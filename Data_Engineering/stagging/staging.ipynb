```python
from faker import Faker
import random
import pandas as pd
import uuid

# Initialize the Faker generator
fake = Faker()

# Function to generate a unique employee ID
def generate_employee_id(index):
    return f"JMD{313 + index}"

# Generate data
def generate_data(num_records):
    data = []
    for i in range(num_records):
        first_name = fake.first_name()
        last_name = fake.last_name()
        email = f"{first_name.lower()}.{last_name.lower()}{i}@example.com"  # Ensures unique email
        phone = fake.phone_number()
        position = random.choice(["Data Scientist", "Data Analyst", "Software Engineer", "Machine Learning Engineer"])
        employeeid = generate_employee_id(i)
        start_date = fake.date_between(start_date='-5y', end_date='today').isoformat()
        date_of_birth = fake.date_of_birth(minimum_age=22, maximum_age=60).isoformat()
        gender = random.choice(["male", "female"])
        address = fake.address()
        employment_status = random.choice(["full_time", "part_time", "contract"])
        user_id = str(uuid.uuid4())  # Generates a unique UUID
        
        record = {
            "firstName": first_name,
            "lastName": last_name,
            "email": email,
            "phone": phone,
            "position": position,
            "employeeid": employeeid,
            "startDate": start_date,
            "dateOfBirth": date_of_birth,
            "gender": gender,
            "address": address,
            "employmentStatus": employment_status,
            "userId": user_id
        }
        
        data.append(record)
    
    return data

# Number of records
num_records = 50000

# Generate data
fake_data = generate_data(num_records)

# Convert to a DataFrame and save as a CSV file (optional)
df = pd.DataFrame(fake_data)
df.to_csv('fake_employee_data.csv', index=False)

print("Data generation completed and saved to 'fake_employee_data.csv'")
```

```python
import pandas as pd
from faker import Faker
import random
import uuid

# Initialize Faker
fake = Faker()

# Function to create unique user IDs
def generate_user_id():
    return str(uuid.uuid4())

# Generate Employee Data
def generate_employee_data(num_records):
    employees = []
    for _ in range(num_records):
        user_id = generate_user_id()
        employee = {
            "firstName": fake.first_name() + " " + fake.last_name(),
            "lastName": fake.last_name(),
            "email": fake.unique.email(),
            "phone": fake.unique.phone_number(),
            "position": random.choice(["Data Scientist", "Software Engineer", "Consultant", "Data Analyst", "Data Engineer"]),
            "employeeid": "JMD" + str(random.randint(1000, 9999)),
            "startDate": fake.date_time_between(start_date='-3y', end_date='now').isoformat(),
            "dateOfBirth": fake.date_of_birth(minimum_age=22, maximum_age=40).isoformat(),
            "gender": random.choice(["male", "female"]),
            "address": fake.address(),
            "employmentStatus": random.choice(["full_time", "part_time", "contract"]),
            "userId": user_id
        }
        employees.append(employee)
    return pd.DataFrame(employees)

# Generate Project Skills Data
def generate_project_skill_data(user_ids):
    project_skills = []
    for user_id in user_ids:
        skill_data = {
            "softwareEngineerProjects": random.randint(0, 10),
            "consultantProjects": random.randint(0, 5),
            "fullStackProjects": random.randint(0, 5),
            "dataAnalystProjects": random.randint(0, 5),
            "dataEngineerProjects": random.randint(0, 10),
            "dataScienceProjects": random.randint(0, 10),
            "otherProjects": random.randint(0, 3),
            "projectType": random.choice(["Data Science", "Software Engineering", "Consulting", "Full Stack"]),
            "toolsUsed": random.choice(["Node.js", "Python", "SQL", "Tableau", "R", "Java"]),
            "favoriteProject": fake.sentence(nb_words=3),
            "userId": user_id
        }
        project_skills.append(skill_data)
    return pd.DataFrame(project_skills)

# Generate Soft Skills Data
def generate_soft_skill_data(user_ids):
    soft_skills = []
    for user_id in user_ids:
        skill_data = {
            "communication": random.randint(1, 10),
            "teamwork": random.choice(["Innovator", "Collaborator", "Leader"]),
            "problemSolving": random.randint(1, 10),
            "adaptability": random.choice(["Visionary", "Flexible", "Resilient"]),
            "timeManagement": random.randint(1, 10),
            "criticalThinking": random.choice(["Leader", "Thinker", "Strategist"]),
            "creativity": random.choice(["Visionary", "Imaginative", "Innovative"]),
            "leadership": random.choice(["Innovator", "Motivator", "Manager"]),
            "interpersonalSkills": random.choice(["Visionary", "Empathic", "Connector"]),
            "emotionalIntelligence": random.choice(["Leader", "Empathic", "Resilient"]),
            "userId": user_id
        }
        soft_skills.append(skill_data)
    return pd.DataFrame(soft_skills)

# Generate Technical Skills Data
def generate_technical_skill_data(user_ids):
    technical_skills = []
    for user_id in user_ids:
        skill_data = {
            "programmingLanguages": random.randint(1, 5),
            "webDevelopment": random.randint(1, 10),
            "databaseManagement": random.randint(1, 10),
            "cloudComputing": random.choice(["Moderately proficient", "Very proficient", "Master"]),
            "versionControl": random.choice(["Moderately proficient", "Very proficient", "Master"]),
            "machineLearning": random.choice(["Beginner", "Intermediate", "Advanced", "Master"]),
            "dataAnalysis": random.choice(["Beginner", "Intermediate", "Advanced", "Master"]),
            "cybersecurity": random.choice(["Moderately proficient", "Very proficient", "Master"]),
            "userId": user_id
        }
        technical_skills.append(skill_data)
    return pd.DataFrame(technical_skills)

# Generate Course Data
def generate_course_data(user_ids, num_records):
    courses = []
    for _ in range(num_records):
        course_data = {
            "courseName": random.choice(["Java", "Python", "Data Science", "Machine Learning", "Web Development"]),
            "platform": random.choice(["Coursera", "edX", "Udemy", "LinkedIn Learning", "ADF"]),
            "specialization": random.choice(["Data Science", "Software Engineering", "Machine Learning", "Web Development"]),
            "totalCoursesCompleted": random.randint(10, 100),
            "courseDuration": random.randint(4, 52),  # Duration in weeks
            "courseType": random.choice(["Online Course", "In-Person Course"]),
            "completionStatus": random.choice(["Completed", "In Progress"]),
            "additionalCourses": fake.word(),
            "verify": random.choice([True, False]),
            "userId": random.choice(user_ids)
        }
        courses.append(course_data)
    return pd.DataFrame(courses)

# Generate Certification Data
def generate_certification_data(user_ids, num_records):
    certifications = []
    for _ in range(num_records):
        cert_data = {
            "certificationTitle": random.choice(["Python", "Machine Learning", "Data Science", "AWS", "Azure"]),
            "platform": random.choice(["Coursera", "edX", "Udemy", "LinkedIn Learning", "Google"]),
            "specialization": random.choice(["Machine Learning", "Cloud Computing", "Data Engineering", "Web Development"]),
            "dateObtained": fake.date_time_between(start_date='-2y', end_date='now').isoformat(),
            "duration": random.randint(10, 100),  # Duration in hours
            "additionalCertifications": fake.word(),
            "competencyLevel": random.choice(["Beginner", "Intermediate", "Advanced"]),
            "userId": random.choice(user_ids)
        }
        certifications.append(cert_data)
    return pd.DataFrame(certifications)

# Generate Achievement Data
def generate_achievement_data(user_ids, num_records):
    achievements = []
    for _ in range(num_records):
        achievement_data = {
            "achievementTitle": fake.job(),
            "organization": fake.company(),
            "dateAchieved": fake.date_time_between(start_date='-3y', end_date='now').isoformat(),
            "description": fake.sentence(nb_words=6),
            "category": random.choice(["Academic", "Professional", "Community Service", "Innovation"]),
            "additionalAchievements": fake.word(),
            "userId": random.choice(user_ids)
        }
        achievements.append(achievement_data)
    return pd.DataFrame(achievements)

# Generate data for 50,000 employees
num_records = 50000
employee_df = generate_employee_data(num_records)

# Extract user IDs for creating related records
user_ids = employee_df['userId'].tolist()

# Generate related datasets
project_skills_df = generate_project_skill_data(user_ids)
soft_skills_df = generate_soft_skill_data(user_ids)
technical_skills_df = generate_technical_skill_data(user_ids)
course_df = generate_course_data(user_ids, 100000)
certification_df = generate_certification_data(user_ids, 100000)
achievement_df = generate_achievement_data(user_ids, 100000)

# Save to CSV files
employee_df.to_csv('employee_data2.csv', index=False)
project_skills_df.to_csv('project_skills_data2.csv', index=False)
soft_skills_df.to_csv('soft_skills_data2.csv', index=False)
technical_skills_df.to_csv('technical_skills_data2.csv', index=False)
course_df.to_csv('courses_data2.csv', index=False)
certification_df.to_csv('certifications_data2.csv', index=False)
achievement_df.to_csv('achievements_data2.csv', index=False)

print("Data generation completed and saved to CSV files.")
```
